### Технологический процесс получения золота из руды  

После добычи руда проходит первичную обработку, после которой получается дробленая смесь. Данную смесь отправляют на флотацию (обогащение) и двухэтапную очистку.  

Опишем каждую стадию: 
1. **Флотация**  

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.  

На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).  

2. **Очистка**   

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Описание данных

Данные находятся в трёх файлах:

Данные индексируются датой и временем получения информации (признак `date`). Соседние по времени параметры часто похожи.  

Некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже. Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков.  

Исходный датасет содержит обучающую и тестовую выборки со всеми признаками.  

**Технологический процесс**  

* *Rougher feed* — исходное сырье
* *Rougher additions* (или *reagent additions*) — флотационные реагенты: *Xanthate*, *Sulphate*, *Depressant*
    * *Xanthate* — ксантогенат (промотер, или активатор флотации);
    * *Sulphate* — сульфат (на данном производстве сульфид натрия);
    * *Depressant* — депрессант (силикат натрия).
* *Rougher process* (англ. «грубый процесс») — флотация
* *Rougher tails* — отвальные хвосты
* *Float banks* — флотационная установка
* *Cleaner process* — очистка
* *Rougher Au* — черновой концентрат золота
* *Final Au* — финальный концентрат золота  

**Параметры этапов**  

* *air amount* — объём воздуха
* *fluid levels* — уровень жидкости
* *feed size* — размер гранул сырья
* *feed rate* — скорость подачи  

**Наименование признаков**  

Наименование признаков:
`[этап].[тип_параметра].[название_параметра]`  

Пример: `rougher.input.feed_ag`  

Возможные значения для блока `[этап]`:
* *rougher* — флотация
* *primary_cleaner* — первичная очистка
* *secondary_cleaner* — вторичная очистка
* *final* — финальные характеристики  

Возможные значения для блока `[тип_параметра]`:
* *input* — параметры сырья
* *output* — параметры продукта
* *state* — параметры, характеризующие текущее состояние этапа
* *calculation* — расчётные характеристики

### Проверка правильности расчета эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке `df_train` для признака `rougher.output.recovery`. Найдем *MAE* между расчётами и значением признака. 

Формула расчета эффективности обогащения:   

$Recovery = \frac{C * (F - T)}{F * (C - T)} * 100 ,$  

где:  
* *С* - доля золота в концентрате после флотации/очистки;
* *F* - доля золота в сырье/концентрате до флотации/очистки;
* *Т* - доля золота в отвальных хвостах после флотации/очистки.  

Определим необходимые признаки для проверки формулы, исходя из полученных данных:  

* `rougher.output.concentrate_au` = *C*
* `rougher.input.feed_au` = *F*
* `rougher.output.tail_au` = *T*  

Посчитанная эффективность находится в значении `rougher.output.recovery`.

```python
target = df_train['rougher.output.recovery']
C = df_train['rougher.output.concentrate_au']
F = df_train['rougher.input.feed_au']
T = df_train['rougher.output.tail_au']

predict = ((C * (F - T)) / (F * (C - T))) * 100

print('MAE между расчетами и значением признака:', MAE(target, predict))
```

**Вывод:**  
* Метрика *МАЕ* (среднее абсолютное отклонение) крайне мала и стремится к нулю, что свидетельствует о правильности расчитанной эффективности обогащения.

### Предобработка данных

Предобработку данных начнем с удаления "лишних признаков" для модели из обучающей выборки. Оставим в обучающей выборке только столбцы, которые присутствуют в тестовой выборке.

Добавим целевые признаки - эффективность обогащения флотации и финальную (`rougher.output.recovery` и `final.output.recovery`) в обучающую и тестовую выборку по дате, после чего удалим столбец с датой из всех выборок, т.к. он не принесет пользы при обучении модели.

```python
# Выделим целевые признаки из исходных данных в отдельный датафрейм
target_signs = df_full[['date', 'rougher.output.recovery', 'final.output.recovery']]

# Добавим целевые признаки в обучающего и тестового датафреймов по дате
df_train = df_train.merge(target_signs, on='date', how='left')
df_test = df_test.merge(target_signs, on='date', how='left')

# Удалим столбец даты из всех датафреймов
df_full = df_full.drop('date', axis=1)
df_train = df_train.drop('date', axis=1)
df_test = df_test.drop('date', axis=1)
```

На этапе изучения исходных файлов выявили, что во всех выборках присутствуют пропуски. Для отображения столбцы с имеющимися пропусками и их количеством напишем функцию:

```python
def isna_columns(df, name_df):
    '''
    Выводит столбцы с пропущенными значениями
    и кол-во пропусков в каждом столбце
    '''
    # Выведем название датафрейма
    print(name_df)
    # Выведем столбцы с пропусками
    display(df.isna().sum().to_frame(name='Пропуски').query('Пропуски > 0'))
```

```python
# Выведем кол-во пропусков в выборках ДО их устранения
isna_columns(df_full, 'Исходная выборка')
isna_columns(df_train, 'Обучающая выборка')
isna_columns(df_test, 'Тестовая выборка')
```

В описании к данным Заказчик указал о том, что соседние значения схожи между собой. Исходя из данной информации, можно сделать вывод, что пропуски лучше всего заполнить методом *k*-ближайших соседей `KNNImputer`. Количество доступных соседей (`n_neighbors`) установим равным 6; весовую функцию (`weights`) установим в значение `distance`, т.е. более близкие соседи точки запроса будут иметь большее влияние, чем соседи, которые находятся дальше.

```python
imputer = KNNImputer(n_neighbors=6, weights='distance')

def imputer_func(df):
    '''
    Заполняет пропуски методом k-ближайших соседей
    '''
    df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
    
    return df

df_full = imputer_func(df_full)
df_train = imputer_func(df_train)
df_test = imputer_func(df_test)
```

```python
# Выведем кол-во пропусков в выборках ПОСЛЕ их устранения
isna_columns(df_full, 'Исходная выборка')
isna_columns(df_train, 'Обучающая выборка')
isna_columns(df_test, 'Тестовая выборка')
```

## Построение модели

### Вычисление итоговой sMAPE

Напишем формулы метрик качества:  

$sMAPE= \frac{1}{N} \sum_{i=1}^{N} \frac {|y_i-\hat{y_i}|}{(|y_i|+|\hat{y_i}|)/2} \times 100\% $

$Final\hspace{0.2cm}sMAPE = 25\% \times sMAPE(rougher) + 75\% \times sMAPE(final)$

Напишем функции для вычисления простой *sMAPE* и итоговой *sMAPE* (*Symmetric Mean Absolute Percentage Error*):

```python
def sMAPE(predictions, target):
    '''
    Вычисляет sMAPE в %
    '''
    smape = np.mean(abs(target - predictions) / ((abs(target) + abs(predictions)) / 2)) * 100
    return smape
```

```python
def final_sMAPE(predictions_rougher, predictions_final, target_rougher, target_final):
    '''
    Вычисляет итоговую sMAPE в %
    '''
    smape_rougher = sMAPE(predictions_rougher, target_rougher)
    smape_final = sMAPE(predictions_final, target_final)
    final_smape = 0.25 * smape_rougher + 0.75 * smape_final
    return final_smape
```

    * Проверили, что в обучающей выборке правильно посчитано значение эффективности обогащения;
    * Проанализировали, каких данных нет в тестовой выборке
    * Обработали данные: удалили "лишние столбцы" из обучающей выборки, добавили целевые признаки в обучающую и тестовую выборки, устранили пропуски во всех выборках методом k-ближайших соседей;
* Проанализировали распределения концентраций металлов на различных этапах обогащения, а также распределения размеров гранул: все распределения имеют нормальный вид. Наблюдались выбросы в районе нулевых значений, которые были удалены из обучающей выборки;
* Подобрали наилучшую модель - `CatBoostRegressor` и сравнили ее значение метрики *sMAPE*, полученной на тестовой выборке с полученным значением константной модели: результат выбранной модели оказался лучше, чем у константной.
